```python
known_genes = adata.var_names
shared_genes = np.intersect1d(known_genes, go_sets.index)
go_sets = go_sets.loc[shared_genes]
min_size = 10
max_size = 100
set_sizes = go_sets.sum()
go_sets = go_sets.loc[:, (set_sizes > min_size) & (set_sizes < max_size)]
```

```python
adata.obs['study_id'] = adata.obs['species'].astype(str)
adata.obs['joint_subclass_label'] = adata.obs['annotation_final_level1B'].astype(str)
```

```python
pymn.MetaNeighbor(adata=adata,
                  study_col='study_id',
                  ct_col='joint_subclass_label',
                  genesets=go_sets,
                  fast_version=True,
                  fast_hi_mem=True)

adata.uns['MetaNeighbor'].to_csv('functional_aurocs.csv')
```
