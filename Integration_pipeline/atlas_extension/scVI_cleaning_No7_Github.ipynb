```python
import sys
import pandas as pd
import numpy as np
import anndata
import scvi
import scanpy as sc
import matplotlib
import os

seed = 10

scvi.settings.seed = 10

sc.logging.print_versions()

```

```python
#celltypes = ["TAL_MD", "TL", "Immune", "PT", "DT", "EC_Stromal", "IC", "PEC"]

celltypes = ["TAL_MD"]

for celltype in celltypes: 
    
    path_to_save_anndata = f'.../Atlas/human_extension/Cleaning_II/Sub_cleaning/{celltype}/{celltype}_subcluster.h5ad'

    adata = sc.read(path_to_save_anndata)
    
    clusters_remove = ["20", "29", "8", "41", "40", "35", "31"]
   
    removed_cells = adata[adata.obs.leiden_scVI_3_0.isin(clusters_remove)].obs_names.unique().tolist()

    adata_subset = adata[~adata.obs.leiden_scVI_3_0.isin(clusters_remove)]

    # Create a DataFrame from the list
    df = pd.DataFrame(removed_cells, columns=["Barcodes"])

    # Save the DataFrame as a CSV file
    df.to_csv(f'.../Atlas/human_extension/Cleaning_II/removed_barcodes/Atlas_human_ext_II_subclustering_{celltype}.csv', index=False)
    
    sc.pl.umap(adata_subset, color = "leiden_scVI_3_0", legend_loc = "on data")
```
